```python
import json
import csv
import time


TOPIK_LIST = [
    "Emosi Negatif dan Suasana Hati", 1
    "Kecemasan dan Overthinking",1
    "Harga Diri dan Identitas Diri",1
    "Relasi Sosial dan Kesepian",1
    "Hubungan Cinta dan Keluarga",1
    "Stres Akademik dan Karier",1
    "Masalah Finansial dan Ketidakpastian Hidup",1
    "Trauma dan Pengalaman Buruk",1
    "Self-Harm dan Pikiran Bunuh Diri",1
    "Pola Hidup Tidak Sehat",1
    "Self-Care dan Pengembangan Diri",1
    "Kebutuhan Bantuan Profesional"1
]

def buat_prompt(question, answer):
    return f"""
Berikut ini adalah pasangan pertanyaan dan jawaban (QnA):

Pertanyaan:
{question}

Jawaban:
{answer}

Dari daftar topik berikut, pilih SATU topik yang paling relevan dengan isi QnA ini:
{chr(10).join(f"- {topik}" for topik in TOPIK_LIST)}

Balas hanya dengan nama topiknya secara persis. Jika tidak cocok, balas dengan "other".
"""

def klasifikasi_dengan_llm(question, answer):
    prompt = buat_prompt(question, answer)
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "Kamu adalah asisten yang bertugas mengklasifikasikan topik psikologis."},
                {"role": "user", "content": prompt}
            ],
            temperature=0
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print("Error:", e)
        return "error"

def proses_label_jsonl(input_path, output_csv_path):
    with open(input_path, "r", encoding="utf-8") as infile, open(output_csv_path, "w", newline='', encoding="utf-8") as outfile:
        writer = csv.DictWriter(outfile, fieldnames=["filename", "question", "answer", "label"])
        writer.writeheader()

        for idx, line in enumerate(infile, start=1):
            data = json.loads(line)
            question = data["question"]
            answer = data["answer"]
            filename = data.get("filename", "unknown")

            label = klasifikasi_dengan_llm(question, answer)
            print(f"{idx}. [{filename}] : {label}")

            writer.writerow({
                "filename": filename,
                "question": question,
                "answer": answer,
                "label": label
            })

            time.sleep(1.2)  
```
